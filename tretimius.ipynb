# Шифр Тритемия
Импорты и алфавит

```python
from colorama import Fore as text_color

alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
```

Функция шифрования

```python
def tritemius(string: str, alphabet: str = alphabet) -> str:
    string = string.upper()
    alphabet = alphabet.upper()
    
    new_str = ''
    for i in range(len(string)):
        new_str += alphabet[(alphabet.index(string[i]) + i) % len(alphabet)]
    return new_str
```

Функция расшифрования

```python
def tritemius_decrypt(string: str, alphabet: str = alphabet) -> str:
    string = string.upper()
    alphabet = alphabet.upper()
    
    new_str = ''
    for i in range(len(string)):
        new_str += alphabet[(alphabet.index(string[i]) - i) % len(alphabet)]
    return new_str
```

Проверка

```python
try:
    open_text = 'НЕВСЕТЕПОВАРАЗПТЧТОСДЛИННЫМИНОЖАМИХОДЯТТЧК'
    encrypted_text = tritemius(open_text)
    decrypted_text = tritemius_decrypt(encrypted_text)

    assert encrypted_text == 'НЖДФЙЧЛЦЦЛКЫМФЭБЗГАДШАЮДЕФЖГЙЛДЯМЙЧСИДШЩЯУ', 'Зашифрованный текст не совпадает с образцом'
    assert open_text == decrypted_text, 'Расшифрованный текст не совпадает с изначальным'
    print(f'{text_color.LIGHTGREEN_EX}Шифрование корректно{text_color.RESET}')
except AssertionError as e:
    print(f'{text_color.LIGHTRED_EX}Шифрование не корректно: {e}{text_color.RESET}')
```
