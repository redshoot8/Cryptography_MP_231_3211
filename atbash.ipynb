# Шифр Атбаш
Импорты и алфавит

```python
from colorama import Fore as text_color
from utils.data_io import get_test_case

alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
```

Функция шифрования

```python
def atbash(string: str, alphabet: str = alphabet) -> str:
    string = string.upper()
    alphabet = alphabet.upper()

    new_str = ''
    for char in string:
        new_str += alphabet[len(alphabet) - alphabet.index(char) - 1] if char in alphabet else char
    return new_str
```

Функция расшифрования

```python
def atbash_decrypt(string: str, alphabet: str = alphabet) -> str:
    string = string.upper()
    alphabet = alphabet.upper()
    new_str = ''
    for char in string:
        if char in alphabet:
            index = len(alphabet) - alphabet.index(char) - 1
            new_str += alphabet[index]
        else:
            new_str += char
    return new_str
```

Проверка

```python
try:
    open_text = 'НЕВСЕТЕПОВАРАЗПТЧТОСДЛИННЫМИНОЖАМИХОДЯТТЧК'
    encrypted_text = atbash(open_text)
    decrypted_text = atbash_decrypt(encrypted_text)
    
    open_text_1000 = get_test_case()
    encrypted_text_1000 = atbash(open_text_1000)
    decrypted_text_1000 = atbash_decrypt(encrypted_text_1000)

    assert encrypted_text == 'ТЪЭОЪНЪРСЭЯПЯШРНИНСОЫФЧТТДУЧТСЩЯУЧКСЫАННИХ', 'Зашифрованный текст не совпадает с образцом'
    assert open_text == decrypted_text, 'Расшифрованный текст не совпадает с изначальным'
    print(f'{text_color.LIGHTGREEN_EX}Шифрование корректно{text_color.RESET}')
except AssertionError as e:
    print(f'{text_color.LIGHTRED_EX}Шифрование не корректно: {e}{text_color.RESET}')
```
