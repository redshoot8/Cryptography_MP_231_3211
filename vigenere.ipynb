# Шифр Виженера
Импорты, алфавит и ключ

```python
from colorama import Fore as text_color
from utils.file_read import get_test_case
from utils.validation import validate_string_key

alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
key = validate_string_key(input('Введите ключ: '), 1)
```

Функция шифрования

```python
def vigenere(string: str, alphabet: str = alphabet, key: str = 'А') -> str:
    string = string.upper()
    alphabet = alphabet.upper()
    
    new_str = ''
    for i in range(len(string)):
        new_str += alphabet[(alphabet.index(string[i]) + alphabet.index(key)) % len(alphabet)]
        key = string[i]
    return new_str
```

Функция расшифрования

```python
def vigenere_decrypt(string: str, alphabet: str = alphabet, key: str = 'А') -> str:
    string = string.upper()
    alphabet = alphabet.upper()
    
    new_str = ''
    for i in range(len(string)):
        new_char = alphabet[(alphabet.index(string[i]) - alphabet.index(key)) % len(alphabet)]
        new_str += new_char
        key = new_char
    return new_str
```

Проверка

```python
try:
    open_text = 'НЕВСЕТЕПОВАРАЗПТЧТОСДЛИННЫМИНОЖАМИХОДЯТТЧК'
    encrypted_text = vigenere(open_text, key=key)
    decrypted_text = vigenere_decrypt(encrypted_text, key=key)

    assert encrypted_text == 'УТЗУЦЧЧФЭРВРРЗЦБЙЙАЯХПУХЪИЗФХЫФЖМФЭГТГСДЙБ', 'Зашифрованный текст не совпадает с образцом'
    assert open_text == decrypted_text, 'Расшифрованный текст не совпадает с изначальным'
    print(f'{text_color.LIGHTGREEN_EX}Шифрование корректно{text_color.RESET}')
except AssertionError as e:
    print(f'{text_color.LIGHTRED_EX}Шифрование не корректно: {e}{text_color.RESET}')
```
