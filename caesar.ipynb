# Шифр Цезаря
Импорты, алфавит и ключ

```python
from colorama import Fore as text_color
from utils.validation import validate_number_key

alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
key = validate_number_key(int(input('Введите ключ: ')))
```

Функция шифрования

```python
def caesar(string: str, alphabet: str = alphabet, key: int = 3) -> str:
    string = string.upper()
    alphabet = alphabet.upper()

    new_str = ''
    for char in string:
        new_str += alphabet[(alphabet.index(char) + key) % len(alphabet)]
    return new_str
```

Функция расшифрования

```python
def caesar_decrypt(string: str, alphabet: str = alphabet, key: int = 3) -> str:
    string = string.upper()
    alphabet = alphabet.upper()
    new_str = ''
    for char in string:
        new_str += alphabet[(alphabet.index(char) - key) % len(alphabet)]
    return new_str
```

Проверка

```python
try:
    open_text = 'НЕВСЕТЕПОВАРАЗПТЧТОСДЛИННЫМИНОЖАМИХОДЯТТЧК'
    encrypted_text_1 = caesar(open_text)
    encrypted_text_2 = caesar(open_text, key=key)
    decrypted_text_1 = caesar_decrypt(encrypted_text_1)
    decrypted_text_2 = caesar_decrypt(encrypted_text_2, key=key)

    assert encrypted_text_1 == 'РИЕФИХИТСЕГУГКТХЪХСФЗОЛРРЮПЛРСЙГПЛШСЗВХХЪН', 'Зашифрованный текст (Ключ 3) не совпадает с образцом'
    assert encrypted_text_2 == 'ХНКЩНЪНЧЦКИШИПЧЪЯЪЦЩМУРХХГФРХЦОИФРЭЦМЗЪЪЯТ', f'Зашифрованный текст (Ключ {key}) не совпадает с образцом'
    assert open_text == decrypted_text_1, 'Расшифрованный текст (Ключ 3) не совпадает с изначальным'
    assert open_text == decrypted_text_2, f'Расшифрованный текст (Ключ {key}) не совпадает с изначальным'
    print(f'{text_color.LIGHTGREEN_EX}Шифрование корректно{text_color.RESET}')
except AssertionError as e:
    print(f'{text_color.LIGHTRED_EX}Шифрование не корректно: {e}{text_color.RESET}')
```

Ввод строки для шифрования
